```python
from langgraph.graph import StateGraph, END
from langchain_core.runnables import RunnableLambda
from langchain_core.documents import Document
import requests
from typing import TypedDict, Dict, Any
from langchain_community.tools.tavily_search import TavilySearchResults
from tavily import TavilyClient
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_groq import ChatGroq
from langchain.chains import RetrievalQA
from dotenv import load_dotenv
import os
load_dotenv()

# Example ChromaDB vectorstore loader
from langchain.vectorstores import Chroma
from langchain.embeddings import HuggingFaceEmbeddings
```

```python
# ====== 1. Define Agent State ======
class AgentState(TypedDict):
    goal: str
    research: str
    audience: str
    strategy: str
    content: str
    simulation: str
    final_plan: str
```

```python
# ====== Initialize LLM (Groq - llama) ======
llm = ChatGroq(model="llama3-70b-8192")
```

```python
#ChromaDB vectorstore loader
def load_chroma_db(path="chroma_db"):
    embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
    vectordb = Chroma(persist_directory=path, embedding_function=embedding_model)
    return vectordb
```

```python
# ======  Define Each Agent ======

# Agent: parse_goal → Groq
def parse_goal(state: AgentState):
    response = llm.invoke(f"Clarify the user's goal: '{state['goal']}'")
    return {"goal": response.content}

# Agent: research_data →  ChromaDB
def research_data(state: AgentState):
    # === ChromaDB similarity search ===
    query = state["goal"]
    vectordb = load_chroma_db("chroma/research")
    chroma_results = vectordb.similarity_search(query, k=5)
    chroma_context = "\n".join([doc.page_content for doc in chroma_results])
    # === 3. Combine both for LLM summarization ===
    prompt = f"""You are an AI assistant helping with marketing and campaign planning.
Given the following goal, summarize 3 key insights from internal chroma data.

Goal:
{state["goal"]}


Internal Data:
{chroma_context}
"""
    llm = ChatGroq(model="llama3-70b-8192")  # Make sure you've imported and set your GROQ_API_KEY
    result = llm.invoke(prompt)

    return {**state, "research": result.content}
# Agent: segment_audience → Groq
def segment_audience(state: AgentState):
    prompt = f"Based on this research: {state['research']}, segment the audience into 2–3 groups."
    result = llm.invoke(prompt)
    return {"audience": result.content}

# Agent: plan_strategy → Groq
def plan_strategy(state: AgentState):
    prompt = f"Given the audience: {state['audience']}, suggest a marketing strategy with 3 key actions."
    result = llm.invoke(prompt)
    return {"strategy": result.content}

# Agent: generate_content → Groq
def generate_content(state: AgentState):
    prompt = f"Using this strategy: {state['strategy']}, write sample campaign messages for each audience segment."
    result = llm.invoke(prompt)
    return {"content": result.content}

# Agent: simulate_campaign → Groq
def simulate_campaign(state: AgentState):
    prompt = f"Estimate campaign performance metrics (discount_percentage,expected_sales,actual_sales etc.) for this content:\n{state['content']}"
    result = llm.invoke(prompt)
    return {"simulation": result.content}

# Agent: export_plan → Plain Text
def export_plan(state: AgentState):
    final = f""" Final Campaign Plan

    Goal:
{state['goal']}

    Research:
{state['research']}

    Audience Segments:
{state['audience']}

    Strategy:
{state['strategy']}

    Campaign Content:
{state['content']}

    Simulated Performance:
{state['simulation']}
"""
    return {"final_plan": final}
```

```python
# ====== 5. Build LangGraph ======
builder = StateGraph(AgentState)

builder.add_node("parse_goal", RunnableLambda(parse_goal))
builder.add_node("research_data", RunnableLambda(research_data))
builder.add_node("segment_audience", RunnableLambda(segment_audience))
builder.add_node("plan_strategy", RunnableLambda(plan_strategy))
builder.add_node("generate_content", RunnableLambda(generate_content))
builder.add_node("simulate_campaign", RunnableLambda(simulate_campaign))
builder.add_node("export_plan", RunnableLambda(export_plan))

# Chain the agents
builder.set_entry_point("parse_goal")
builder.add_edge("parse_goal", "research_data")
builder.add_edge("research_data", "segment_audience")
builder.add_edge("segment_audience", "plan_strategy")
builder.add_edge("plan_strategy", "generate_content")
builder.add_edge("generate_content", "simulate_campaign")
builder.add_edge("simulate_campaign", "export_plan")
builder.add_edge("export_plan", END)

graph = builder.compile()
```

```python
# ====== 6. Run the Agentic AI ======
initial_input = {
    "goal": "Create a campaign to boost SUV sales in North region of India.",
    "research": "SUV sales have increased in urban areas; North region underperforming.",
    "audience": "Young professionals,Budget Buyers,Luxury Car owners in urban North India.",
    "strategy": "Focus on social media channels and festive discounts.",
    "content": "Catchy videos + influencer posts highlighting SUV features.",
    "simulation": "Expected 25% lift in leads over 3 weeks.",
     "final_plan": ""
}

result = graph.invoke(initial_input)

print("\n Final Campaign Plan:\n")
print(result["final_plan"])
```
